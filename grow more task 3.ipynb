### Converting the original image to  greyscale image

```python
import cv2
import matplotlib.pyplot as plt
img_grey=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img_grey)
```

### Inverting the greyscale image(negative image)

```python
invert=cv2.bitwise_not(img_grey)
plt.imshow(invert)
```

### Blurry image

```python
blur=cv2.GaussianBlur(invert,(21,21),0)
plt.imshow(blur)
```

### Inverted blurry image

```python
invertedblur=cv2.bitwise_not(blur)
plt.imshow(invertedblur)
```

### Obtaining a sketch

```python
sketch=cv2.divide(img_grey,invertedblur,scale=200.0)
plt.imshow(sketch)
```

### Final sketch

```python
image=cv2.cvtColor(sketch,cv2.COLOR_RGB2BGR)
plt.imshow(image)
plt.axis(False)
plt.show()
```
